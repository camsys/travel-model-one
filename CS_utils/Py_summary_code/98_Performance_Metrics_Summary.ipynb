```python
metric_name = []
metric_val = []
total_values = []
special_case = []
category_name = []
category_val = []
actual_perceived = []
all_filter_conditions = []
description = []

files = os.listdir(os.getcwd())
for new_file in files:
    split_file_name = new_file.split('_') #To identiy the metric

    for metric in list(dict_of_metrics.keys()): #Gathers the list of metrics defined in above dictionary
        if str(metric) in split_file_name: #Read the file if listed metric is found
            df = pd.read_csv(new_file)

            # Check if the file name mentions as actual or perceived
            if 'actual' in new_file:
                actual_perceived_val = 'actual'
            elif 'perceived' in new_file:
                actual_perceived_val = 'perceived'
            else:
                actual_perceived_val = ''

            # Extract the performance metric through sequential filtering of the data
            filter_conditions_this_metric = []
            for key in list(dict_of_metrics[metric].keys()): #Gathers the defined filters for each metric
                condition = key
                filter_key = dict_of_metrics[metric][key]
                df = df[df[condition]==filter_key]
                filter_conditions_this_metric.append((condition, filter_key))
            output_values = df['Value'].values

            if df.shape[0] > 0:
                description_val = df.Description.values[0]

            # Condition for metric if it need not be summarized by categories such as 'Zone_ID', 'Mode'
            if metric not in list(summarize_by.keys()):
                if len(output_values) == 0:
                    break
                elif len(output_values)>1:
                    if metric in max_agg_of_metrics:
                        final_output = max(output_values)
                    else:
                        final_output = sum(output_values)
                elif metric in divide_by_pop:
                    final_output = output_values[0]/total_population
                else:
                    final_output = output_values[0]

                metric_val.append('{0:.2f}'.format(final_output))
                metric_name.append(metric)
                total_values.append(len(output_values))
                category_name.append('')
                category_val.append('')
                actual_perceived.append(actual_perceived_val)
                all_filter_conditions.append(filter_conditions_this_metric)
                description.append(description_val)
            
            # Condition for metric if it must summarized by categories such as 'Zone_ID', 'Mode' as listed in 'summarize_by' dictionary
            if metric in list(summarize_by.keys()):

                for index, row in df.iterrows():
                    metric_val.append('{0:.2f}'.format(row['Value']))
                    metric_name.append(metric)
                    total_values.append(1)
                    category_name.append(summarize_by[metric])
                    category_val.append(row[summarize_by[metric]])
                    actual_perceived.append(actual_perceived_val)
                    all_filter_conditions.append(filter_conditions_this_metric)
                    description.append(description_val)
                                 
```
